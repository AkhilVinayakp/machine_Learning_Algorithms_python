```python
# As the year sold increases Sale price decreases
# As the Year build increase Sale price also increase
# calculate the age of the house and what is the pattern of price
# th age can be calculated from the Year build Year of renovation basically any other in temporal group
```

```python
df['age from sold'] = df['YrSold'] - df['YearBuilt']
df_temp = df.groupby("YrSold")[['YrSold', 'age from sold', 'SalePrice']].median()
```

```python
df_temp
```

```python


fig, ax1 = plt.subplots()
x = df_temp['YrSold']
y1 = df_temp['SalePrice']
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('Year sold')
ax1.set_ylabel('Sale Price', color='g')  # Y-axis label for the first plot
ax1.tick_params(axis='y', labelcolor='g')  # Y-axis ticks for the first plot

# Create a second set of axes sharing the same x-axis
ax2 = ax1.twinx()
y2 = df_temp['age from sold']
# Plot data on the second y-axis (right)
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Age from Sold', color='b')  # Y-axis label for the second plot
ax2.tick_params(axis='y', labelcolor='b')  # Y-axis ticks for the second plot

# Title for the entire plot
plt.title('Age vs Sale price')

# Show the plot
plt.show()
```
