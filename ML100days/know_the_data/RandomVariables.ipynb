```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the range of x values and the PDF of a normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  # Standard normal distribution

# Specific point for Pr(X = c)
c = 1  # Example value
pdf_c = norm.pdf(c, loc=0, scale=1)

# Interval example for Pr(a <= X <= b)
a, b = 0, 2  # Example interval
x_interval = np.linspace(a, b, 500)
pdf_interval = norm.pdf(x_interval, loc=0, scale=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Probability Density Function (PDF)", color='blue')
plt.fill_between(x_interval, pdf_interval, color='skyblue', alpha=0.5, label=f"Pr({a} ≤ X ≤ {b})")
plt.scatter(c, pdf_c, color='red', label=f"Pr(X = {c}) = 0", zorder=5)
plt.axvline(c, color='red', linestyle='--', alpha=0.7)

# Labels and legend
plt.title("Continuous Random Variable and Probability", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Density (f(x))", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
