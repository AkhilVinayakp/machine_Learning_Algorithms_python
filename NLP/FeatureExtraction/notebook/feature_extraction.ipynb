```python
prompt_template = """
You are a luxury real estate assistant.
Your task is to extract the unique and standout maximum 5 features  of a luxury home from the listing remarks below.

Listing: {remarks}

Please output the special features in the following JSON format:
{{
    "best_features": [
        "feature1",
        "feature2",
        "feature3"
    ]
}}

Ensure the response is valid JSON with no extra text or explanations.
"""

# Sample listing remarks
listing_remarks = df['listingRemarks'][345]

# Format the prompt with listing details
formatted_prompt = prompt_template.format(remarks=listing_remarks)

# Send the request to Deepseek via Ollama
data = {
    "model": "deepseek-r1:1.5b",
    "prompt": formatted_prompt,
    "stream": False
}

import requests

url = "http://localhost:11434/api/generate"
response = requests.post(url, json=data)
response_text = response.json().get("response")
clean_output = re.sub(r"<think>.*?</think>", "", response_text, flags=re.DOTALL)
# Print the model's response
print("Input text >>>", listing_remarks)
print("response text as")
print(response_text)
print("cleaned output")
print(clean_output)
```
