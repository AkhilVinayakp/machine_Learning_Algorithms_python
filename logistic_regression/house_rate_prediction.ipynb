```python
import turicreate
```

```python
# loading the data
data = turicreate.SFrame("home_data.sframe")
```

```python
data
```

```python
# ploting the data
plt = turicreate.visualization.scatter(data["sqft_living"], data["price"])
```

```python
train_data, test_data = data.random_split(.8, seed=0, exact=True)
```

```python
turicreate.visualization.scatter(test_data['sqft_living'],test_data['price'],xlabel="sqrt", ylabel='price')
```

```python
# creating the model using turicreate 
lin_mod = turicreate.linear_regression.create(train_data, target='price', features=['sqft_living'] )
```

```python
# evaluating the model 
lin_mod.evaluate(test_data)
```
