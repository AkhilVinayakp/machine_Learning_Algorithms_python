```python
import turicreate
import numpy as np
import json
import string
```

```python
# loading the data
products = turicreate.SFrame('amazon_baby_subset.sframe')
```

```python
with open('important_words.json','r') as f:
    important_words = json.load(f)
```

```python
# removing the punctuations
def remove_punch(t):
    translater = t.maketrans('','',string.punctuation)
    return t.translate(translater)
```

```python
products['cl_review'] = products['review'].apply(remove_punch)
```

```python
products
```

```python
for w in important_words:
    products[w] = products['cl_review'].apply(lambda s:s.split().count(w))
```

```python
products
```

```python
products['con_perfect'] = products['perfect'].apply(lambda c:1 if c >= 1 else 0)
```

```python
# constructing the numpy array
```

```python
def get_numpy_data(data_sframe, features, label):
    data_sframe['intercept'] = 1
    features = ['intercept'] + features
    features_sframe = data_sframe[features]
    feature_matrix = features_sframe.to_numpy()
    label_sarray = data_sframe[label]
    label_array = label_sarray.to_numpy()
    return(feature_matrix, label_array)
```

```python
feature_matrix, sentiment = get_numpy_data(products, important_words, 'sentiment') 
```
