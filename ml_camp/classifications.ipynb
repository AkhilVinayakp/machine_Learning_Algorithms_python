```python
from sklearn.linear_model import LogisticRegression
from  sklearn.datasets import load_breast_cancer
```

```python
cancer  = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
X = scale(X)
```

```python
m_l1 = LogisticRegression(penalty='l1',solver='liblinear')
```

```python
m_l2 = LogisticRegression()
```

```python
m_l1.fit(X,y)
m_l2.fit(X,y)
```

```python
# multiclass classification using multinomial
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
wine= load_wine()
X = wine.data
y = wine.target
print(X.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size = .4, random_state = 42)
```

```python
lr_one_vs_rest = LogisticRegression()
lr_multinomial = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
lr_one_vs_rest.fit(x_train,y_train)
lr_multinomial.fit(x_train,y_train)
print(f"oneVSrestScore{lr_one_vs_rest.score(x_test,y_test)}")
print(f"multi{lr_multinomial.score(x_test,y_test)}")
print('trainig acc')
print(f"oneVSrestScore{lr_one_vs_rest.score(x_train,y_train)}")
print(f"multi{lr_multinomial.score(x_train,y_train)}")
```
